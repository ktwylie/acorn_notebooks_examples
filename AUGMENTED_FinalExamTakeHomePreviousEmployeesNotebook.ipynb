```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
import numpy as np
import pandas as pd
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
```

```python
# So many of these creatures are the species ALLIORATED, I could write a classifier that just 
# consisted of a single print statement that predicts "ALLIORATED" for every specimen and 
# save myself a lot of time!

slf_dataset = pd.read_csv('with_cdd_FinalExamTakeHomeSulfurLifeform.csv').values

labelencoder = LabelEncoder()
slf_dataset[:,-1] = labelencoder.fit_transform(slf_dataset[:,len(slf_dataset[0])-1])

data = slf_dataset[:,:-1] #all columns except the last one
target = slf_dataset[:,len(slf_dataset[0])-1] #only the last column

target = target.astype(int)

min_max_scaler = preprocessing.MinMaxScaler()
data = min_max_scaler.fit_transform(data)

Xtrain, xtest, Ytrain, ytest = train_test_split(data, target, random_state=432)

model=0
```

```python
#Question 3 analysis. 
test = pd.read_csv('FinalExamTakeHomeSulfurLifeform.csv') #Load in new dummy version of the database. 
test.set_index(["AP", "HEV", "Species"]).count(level="Species") #Counts occurances of each species. 
#test
```

```python
# Builds Nearest Neighbor Classifier to classify the Sulfur Life Forms into Species
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(Xtrain, Ytrain)
knn.score(xtest, ytest)
```

```python
# Builds Neural Network to classify the Sulfur Life Forms into Species 
def buildAndTrainModel(numEpochs) :
    global model, Xtrain, Ytrain, xtest, ytest
    model = Sequential()
    model.add(Dense(16, input_dim=4))
    model.add(Activation('tanh'))
    model.add(Dense(16))
    model.add(Activation('tanh'))
    model.add(Dense(5))
    model.add(Activation('softmax')) # Use sigmoid activation function

    # create a Stochastic Gradient Descent (SGD) optimizer with a learning rate of 0.1
    sgd = SGD(lr=0.35)
    # build the model NN as described by the above commands
    model.compile(loss='mean_squared_error', optimizer=sgd,metrics = ["accuracy"])
    
    # do a user specified number of iterations of forward and backward propagation to train the network
    # all operations here are done behind the scenes, all updating of weights and
    # calculation of errors
    model.fit(Xtrain, Ytrain, batch_size=1, epochs=numEpochs)

# Predicts the results for test data set. First gives the input data 
def predictResults () :
    sum=0
    for i in range(len(xtest)) :
        results = model.predict_proba(np.array([xtest[i]]))
        result = np.argmax(results[0])  # argmax does the same thing as PosOfMaxOfThree
        if (result == np.argmax(ytest[i])):  # except that it works for any number of values
            sum += 1
    print("Accuracy:",(sum/len(ytest)) * 100)

#############################################################################################
    
Ytrain = to_categorical(Ytrain) # Only do this once!

buildAndTrainModel(50)

ytest = to_categorical(ytest) # Only do this once!

predictResults()
```
